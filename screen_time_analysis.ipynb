```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\HP\Downloads\screentime_details.csv")
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
# Usage of apps over time
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Date', y='Usage', hue='App')
plt.title('Screen Time Analysis: Usage')
plt.xlabel('Date')
plt.ylabel('Usage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Number of notifications from apps over time
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Date', y='Notifications', hue='App')
plt.title('Screen Time Analysis: Notifications')
plt.xlabel('Date')
plt.ylabel('Notifications')
plt.xticks(rotation=45)
plt.show()
```

```python
# Number of times apps opened over time
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Date', y='Times opened', hue='App')
plt.title('Screen Time Analysis: Times Opened')
plt.xlabel('Date')
plt.ylabel('Times Opened')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship between number of notifications and usage
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Notifications', y='Usage', size='Notifications')
plt.title('Relationship Between Number of Notifications and Usage')
plt.xlabel('Number of Notifications')
plt.ylabel('Usage')
plt.show()
```

```python
# Box plot of app usage
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='App', y='Usage')
plt.title('Screen Time Analysis: App Usage')
plt.xlabel('App')
plt.ylabel('Usage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
correlation = data[['Usage', 'Notifications', 'Times opened']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairwise scatter plot
plt.figure(figsize=(10, 6))
sns.pairplot(data=data, vars=['Usage', 'Notifications', 'Times opened'], hue='App')
plt.title('Pairwise Scatter Plot')
plt.show()
```

```python
# Violin plot of app usage
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='App', y='Usage')
plt.title('Screen Time Analysis: App Usage')
plt.xlabel('App')
plt.ylabel('Usage')
plt.xticks(rotation=45)
plt.show()
```
