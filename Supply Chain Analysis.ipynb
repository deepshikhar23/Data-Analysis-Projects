```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\HP\Downloads\supplychain\supply_chain_data.csv")
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
# Relationship between price and revenue generated
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Price', y='Revenue generated', hue='Product type')
plt.title('Supply Chain Analysis: Price vs Revenue Generated')
plt.show()
```

```python
# Sales by product type
sales_data = data.groupby('Product type')['Number of products sold'].sum().reset_index()

plt.figure(figsize=(8, 6))
plt.pie(sales_data['Number of products sold'], labels=sales_data['Product type'], autopct='%1.1f%%')
plt.title('Sales by Product Type')
plt.show()
```

```python
# Total revenue by shipping carrier
total_revenue = data.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=total_revenue, x='Shipping carriers', y='Revenue generated')
plt.title('Total Revenue by Shipping Carrier')
plt.xlabel('Shipping Carrier')
plt.ylabel('Revenue Generated')
plt.show()
```

```python
# Average lead time and average manufacturing costs
avg_lead_time = data.groupby('Product type')['Lead time'].mean().reset_index()
avg_manufacturing_costs = data.groupby('Product type')['Manufacturing costs'].mean().reset_index()
result = pd.merge(avg_lead_time, avg_manufacturing_costs, on='Product type')
result.rename(columns={'Lead time': 'Average Lead Time', 'Manufacturing costs': 'Average Manufacturing Costs'}, inplace=True)
print(result)
```

```python
# Revenue generated by SKU
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='SKU', y='Revenue generated')
plt.title('Revenue Generated by SKU')
plt.show()
```

```python
# Stock levels by SKU
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='SKU', y='Stock levels')
plt.title('Stock Levels by SKU')
plt.show()
```

```python
# Order quantity by SKU
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='SKU', y='Order quantities')
plt.title('Order Quantity by SKU')
plt.show()
```

```python
# Shipping costs by carrier
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Shipping carriers', y='Shipping costs')
plt.title('Shipping Costs by Carrier')
plt.xlabel('Shipping Carrier')
plt.ylabel('Shipping Costs')
plt.show()
```

```python
# Average defect rates by product type
defect_rates_by_product = data.groupby('Product type')['Defect rates'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=defect_rates_by_product, x='Product type', y='Defect rates')
plt.title('Average Defect Rates by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Defect Rates')
plt.show()
```

```python
# Defect rates by transportation mode
pivot_table = pd.pivot_table(data, values='Defect rates', index=['Transportation modes'], aggfunc='mean')

plt.figure(figsize=(8, 6))
plt.pie(pivot_table["Defect rates"], labels=pivot_table.index, autopct='%1.1f%%')
plt.title('Defect Rates by Transportation Mode')
plt.show()
```
